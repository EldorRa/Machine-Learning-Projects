```python
import numpy as numpy
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/nlp-getting-started/train.csv")
test_data = pd.read_csv("/content/drive/MyDrive/nlp-getting-started/test.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
X_train = TabularDataset(train_data)
X_test = TabularDataset(test_data)
```

```python
model = TabularPredictor("target", eval_metric="f1").fit(X_train,time_limit=60,presets="best_quality",)
```

```python
data_preds = pd.DataFrame()
data_preds["id"] = test_data["id"]
data_preds["target"] = model.predict(X_test)
data_preds
```

```python
data_preds.to_csv("/content/drive/MyDrive/nlp-getting-started/test_preds.csv", index = False)
```
